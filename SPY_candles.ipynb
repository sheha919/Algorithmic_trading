```python
#plot data
mpf.plot(df)
```

```python
# plot as a line curve with volume
mpf.plot(df, type='line', volume=True)
```

```python
#plot data for 2020-09
mpf.plot(df['2020-09-01':'2020-10-01'], volume =True)
```

```python
#plot candlestick chart
mpf.plot(df['2020-09-01':'2020-10-01'], type = 'candle', volume=True)
```

```python
#add moving average to the candlestick chart
mpf.plot(df['2020-09':'2020-12'], type = 'candle', volume=True, mav=(10), tight_layout=True)
```

```python
mpf.plot(df['2020-09':'2020-12'], type = 'candle', volume=True, mav=(10), 
         figratio= (20,12),
         title= 'SPY Price Sep-Dec 2020',
         tight_layout=True,
        style= 'yahoo' )
```
